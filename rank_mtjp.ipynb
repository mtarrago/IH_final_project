```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
url = "https://morethanjustparks.com/national-parks-ranked/"
    
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())  # This formats the HTML in a readable way
```

```python
scraped = soup
```

```python
scraped
```

```python
h3_elements = scraped.find_all(name='h3',class_="wp-block-heading")
list_elements = []
for element in h3_elements:
    if 'has-text-align' not in str(element) and 'Premium' not in str(element) and 'Helpful' not in str(element):
        list_elements.append(str(element))

new_list = []
for i in list_elements:
    i = str(i).split('. ')
    new_list.append(i)

new_list2 = []
for i in new_list:
    i = str(i[1]).split('<')
    new_list2.append(i[0])
```

```python
new_list2
```

```python
final = pd.DataFrame({'rank_mtjp':(np.arange(1,len(new_list2)+1)),'parks_ranked':new_list2})
```

```python
final.to_csv('rank_mtjp.csv')
```
